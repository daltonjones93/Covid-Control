```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python

G0 = nx.Graph()
G1 = nx.Graph()
G2 = nx.Graph()
G3 = nx.Graph()

Gn = [G0,G1,G2,G3]

n1 = 17
p1 = .12

deg_dist1 = binom.pmf(np.arange(n1),n1,p1)
deg_dist1/=np.sum(deg_dist1)

n2 = 11
p2 = .32

deg_dist2 = binom.pmf(np.arange(n2),n2,p2)
deg_dist2/=np.sum(deg_dist2)

n3 = 25
p3 = .2

deg_dist3 = binom.pmf(np.arange(n3),n3,p3)
deg_dist3/=np.sum(deg_dist3)

n4 = 25
p4 = .3

deg_dist4 = binom.pmf(np.arange(n4),n4,p4)
deg_dist4/=np.sum(deg_dist4)


    
deg_dist = [deg_dist1,deg_dist2,deg_dist3,deg_dist4]


#let's create some graphs...



for ii in range(4):
    dist = deg_dist[ii]
    
    if ii == 0:
        print(dist)
    for i in range(20):
        #sample degree:
        
        deg= int(np.random.choice(len(dist),p = dist))
        
        edges = np.random.choice(150,deg,replace = False)
        for j in edges:
            Gn[ii].add_edge(i,j)

            



```

```python
fig, axes = plt.subplots(nrows=2, ncols=2,figsize = (8,8))
ax = axes.flatten()

# for i in range(4):
#     nx.draw_networkx(G[i], ax=ax[i],node_size = 40,with_labels=False)
#     ax[i].set_xlabel('Test',fontsize = 15)
# #     ax[i].set_axis_off()

nx.draw_networkx(G0, ax=ax[2],node_size = 40,with_labels=False)
ax[2].set_xlabel('Low '+r'$\sigma_x^2$',fontsize = 30)
ax[2].set_ylabel('Low '+r'$\mu_x$',fontsize = 30)

nx.draw_networkx(G1, ax=ax[3],node_size = 40,with_labels=False)
ax[3].set_xlabel('High '+r'$\sigma_x^2$',fontsize = 30)

nx.draw_networkx(G2, ax=ax[0],node_size = 40,with_labels=False)
# ax[0].set_xlabel('lowsighighmu',fontsize = 15)
ax[0].set_ylabel('High '+r'$\mu_x$',fontsize = 30)

nx.draw_networkx(G3, ax=ax[1],node_size = 40,with_labels=False)
# ax[1].set_xlabel('highsighighmu',fontsize = 15)




fig.suptitle('Binomial Distribution',fontsize = 35,x = .57,y=1.045)
plt.tight_layout()
plt.savefig('BinomialGraphsBigFont.pdf',bbox_inches = 'tight')
plt.show()
```
