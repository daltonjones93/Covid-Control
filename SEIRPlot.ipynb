```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')

```

```python
def SEIRPlot(I=.01,S= .99, R =0,stepsize = .007,k=1,beta = 10, steps = 1000):
    Is = [np.copy(I)]
    Ss = [np.copy(S)]
    Rs = [np.copy(R)]
    
    for i in range(steps):
        dI = beta*S*I-k*I
        dR = k*I
        dS = -beta*S*I
        
        I += stepsize*dI
        R += stepsize*dR
        S += stepsize*dS
        Is.append(I)
        Rs.append(R)
        Ss.append(S)
    
    plt.figure(figsize = (4,3))
    plt.plot(Is,label = 'Infected')
    plt.plot(Rs,label = 'Recovered')
    plt.plot(Ss,label = 'Susceptible')
    plt.legend()
    plt.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=False)
    plt.savefig('mit-beamer-master/SEIRModel.png')
    plt.show()
    
    
```

```python
SEIRPlot()
```
